## requests + BeautifulSoup


```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.youtube.com/'
r = requests.get(url)
r.encoding = 'utf-8'
soup = BeautifulSoup(r.text)
```

```python
print(soup.prettify())
```

```python
print(type(soup.title))
print(soup.title)
print(soup.title.text)
```

## Grab + PyQuery

```python
from grab import Grab
from pyquery import PyQuery as pq
g = Grab()
resp=g.go(url)
doc = pq(resp.body)
```

```python
title = doc('title')
print(type(title))
print(title)
```

```python
#可直接用PyQuery解析
p = pq(url)

print(type(p('title')))
print(p('title'))
```
